```python
# write app.py file for streamlit
%%writefile app.py

import streamlit as st
import pandas as pd
import sqlite3
from sqlalchemy import create_engine,text

#Creating a database called Census in sqlitedb
path='sqlite:///Census.db'

#Connection for database
my_con=create_engine(path).connect()

#Q1. What is the total population of each district?

df1=pd.read_sql(text('select District,Population from Census_2011'),my_con)
st.write('Total population of each district')
st.write(df1)

#Q2. How many literate males and females are there in each district?

df1=pd.read_sql(text('select District,Literate_Male,Literate_Female from Census_2011'),my_con)
st.write('Total literate males and females are there in each district')
st.write(df1)

#Q3. What is the percentage of workers (both male and female) in each district?

df1=pd.read_sql(text('''select District,Workers,Male_Workers,Female_Workers,
(Male_Workers/Workers)*100 as Male_Worker_Percent,
(Female_Workers/Workers)*100 as Female_Worker_Percent from Census_2011'''),my_con)
st.write('Percentage of workers (both male and female) in each district')
st.write(df1)

#Q4. How many households have access to LPG or PNG as a cooking fuel in each district?

df1=pd.read_sql(text('select District,LPG_or_PNG_Households from Census_2011'),my_con)
st.write(' No. of households have access to LPG or PNG as a cooking fuel in each district')
st.write(df1)

#Q5. What is the religious composition (Hindus, Muslims, Christians, etc.) of each district?

df1=pd.read_sql(text('''select District,(Hindus/Population)*100 as Hindus_Percent,
(Muslims/Population)*100 as Muslims_Percent,
(Christians/Population)*100 as Christians_Percent,(Sikhs/Population)*100 as Sikhs_Percent,
(Buddhists/Population)*100 as Buddhists_Percent,(Jains/Population)*100 as Jains_Percent,
(Others_Religions/Population)*100 as Others_Religions_Percent,
(Religion_Not_Stated/Population)*100 as Religion_Not_Stated_Percent  from Census_2011'''),my_con)
st.write('Religious composition (Hindus, Muslims, Christians, etc.) of each district')
st.write(df1)

#Q6. How many households have internet access in each district?

df1=pd.read_sql(text('select District,Households_with_Internet from Census_2011'),my_con)
st.write('No. of households have internet access in each district')
st.write(df1)

#Q7. What is the educational attainment distribution (below primary, primary, middle, secondary, etc.) in each district?

df1=pd.read_sql(text('''select District, Below_Primary_Education, Primary_Education, Middle_Education, Secondary_Education,
Higher_Education, Graduate_Education, Other_Education from Census_2011'''),my_con)
st.write('Educational attainment distribution (below primary, primary, middle, secondary, etc.) in each district')
st.write(df1)

#Q8. How many households have access to various modes of transportation (bicycle, car, radio, television, etc.) in each district?

df1=pd.read_sql(text('''select District, Households_with_Bicycle, Households_with_Car_Jeep_Van,
Households_with_Scooter_Motorcycle_Moped, Households_with_all from Census_2011'''),my_con)
st.write('''No. of households have access to various modes of
transportation (bicycle, car, radio, television, etc.) in each district''')
st.write(df1)

#Q9. What is the condition of occupied census houses (dilapidated, with separate kitchen, with bathing facility, with latrine facility, etc.) in each district?

df1=pd.read_sql(text('''select District,Occupied_census_houses_Dilapidated_Households,
Households_with_kitchen_Cooking_inside_house, Having_bathing_facility_Total_Households,
Having_latrine_facility_within_the_premises from Census_2011'''),my_con)
st.write('''Condition of occupied census
  houses (dilapidated, with separate kitchen, with bathing facility, with latrine facility, etc.) in each district''')
st.write(df1)

#Q10. How is the household size distributed (1 person, 2 persons, 3-5 persons, etc.) in each district?

df1=pd.read_sql(text('''select District, Household_size_1_person_Households, Household_size_2_persons_Households,
Household_size_1_to_2_persons, Household_size_3_persons_Households, Household_size_3_to_5_persons_Households,
Household_size_4_persons_Households, Household_size_5_persons_Households, Household_size_6_8_persons_Households,
Household_size_9_persons_and_above_Households from Census_2011'''),my_con)
st.write('Household size distributed (1 person, 2 persons, 3-5 persons, etc.) in each district')
st.write(df1)

#Q11. What is the total number of households in each state?

df1=pd.read_sql(text('select `State/UT` , sum(Households) as Total_Household from census_2011 group by `State/UT`'),my_con)
st.write('Total number of households in each state')
st.write(df1)

#Q12. How many households have a latrine facility within the premises in each state?

df1=pd.read_sql(text('select `State/UT` , sum(Having_latrine_facility_within_the_premises) as Total_households_having_latrine_facility from census_2011 group by `State/UT`'),my_con)
st.write('No. of households have a latrine facility within the premises in each state')
st.write(df1)

#Q13. What is the average household size in each state?

df1=pd.read_sql(text('select `State/UT` , avg(Households) as Average_Household_size from census_2011 group by `State/UT`'),my_con)
st.write('Average household size in each state')
st.write(df1)

#Q14. How many households are owned versus rented in each state?

df1=pd.read_sql(text('''select `State/UT`, sum(Ownership_Owned_Households), sum(Ownership_Rented_Households)
 from census_2011 group by `State/UT`'''),my_con)
st.write('No. of households are owned versus rented in each state')
st.write(df1)

#Q15. What is the distribution of different types of latrine facilities (pit latrine, flush latrine, etc.) in each state?

df1=pd.read_sql(text('''select `State/UT` , sum(Type_of_latrine_facility_Pit_latrine_Households),
sum(Type_of_latrine_facility_Other_latrine_Households), sum(Type_of_latrine_facility_open_drain),
sum(Type_of_latrine_flush_latrine) from census_2011 group by `State/UT`'''),my_con)
st.write('Disrtibution of different types of latrine facilities (pit latrine, flush latrine, etc.) in each state')
st.write(df1)

#Q16. How many households have access to drinking water sources near the premises in each state?

df1=pd.read_sql(text('select `State/UT` ,sum(Location_of_drinking_water_source_Near_the_premises) from census_2011 group by `State/UT`'),my_con)
st.write('No. of households have access to drinking water sources near the premises in each state')
st.write(df1)

#Q17. What is the average household income distribution in each state based on the power parity categories?

df1=pd.read_sql(text('''select `State/UT` ,avg(Power_Parity_Less_than_Rs_45000), avg(Power_Parity_Rs_45000_90000),
 avg(Power_Parity_Rs_90000_150000), avg(Power_Parity_Rs_45000_150000), avg(Power_Parity_Rs_150000_240000),
 avg(Power_Parity_Rs_240000_330000), avg(Power_Parity_Rs_150000_330000), avg(Power_Parity_Rs_330000_425000),
 avg(Power_Parity_Rs_425000_545000), avg(Power_Parity_Rs_330000_545000), avg(Power_Parity_Above_Rs_545000) from census_2011 group by `State/UT`'''),my_con)
st.write('Average household income distribution in each state based on the power parity categories')
st.write(df1)

#Q18. What is the percentage of married couples with different household sizes in each state?

df1=pd.read_sql(text('''select `State/UT` ,Married_couples_1_Households, Married_couples_2_Households,
Married_couples_3_Households, Married_couples_3_or_more_Households, Married_couples_4_Households,
Married_couples_5__Households, Married_couples_None_Households from census_2011 group by `State/UT`'''),my_con)
st.write('Percentage of married couples with different household sizes in each state')
st.write(df1)

#Q19. How many households fall below the poverty line in each state based on the power parity categories?
df1=pd.read_sql(text('select `State/UT` , sum(Power_Parity_Less_than_Rs_45000) from census_2011 group by `State/UT`'),my_con)
st.write('No. of households fall below the poverty line in each state based on the power parity categories')
st.write(df1)

#Q20. What is the overall literacy rate (percentage of literate population) in each state?

df1=pd.read_sql(text('''select `State/UT` , Population,Literate, sum((Literate/Population)*100) as Literate_Population
from census_2011 group by `State/UT`'''),my_con)
st.write('Overall literacy rate (percentage of literate population) in each state')
st.write(df1)
```
